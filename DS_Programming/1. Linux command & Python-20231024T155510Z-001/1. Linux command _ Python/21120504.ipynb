- After checking the data, design a function named `read_data` with the name of data file as parameter, which returns a dictionary. Each element of this dictionary follows:
    - Key: Name of the columns
    - Value: Lists which contains the data of the coresponding columns

````python

def read_data(file_name):
    # YOUR CODE HERE

    file = open(file_name, 'r')
    # read the first line as column headers
    headers = file.readline().strip().split(',')

    result = {header: [] for header in headers}

    for line in file:
        values = line.strip().split(',')
        for i, header in enumerate(headers):
            result[header].append(values[i])
    
    return result
    #raise NotImplementedError()


data_2018 = read_data('./data/data_2018.csv')
data_2019 = read_data('./data/data_2019.csv')
````

### 3.2. Duplicated columns

- Observe the data in 2018, you can see that columns `Accident Fields_Reference Number`, `Vehicle Fields_Reference Number` and `Casualty Fields_Reference Number` seem to contain quite similar data.

- Additionally, columns `Number of Vehicles` and `Vehicle Number` in `data_2018` and `data_2019` also look quite similar.

- In order to make sure of these hypothesises, you have to check these columns. Your mission is to design a function named `check_similarity` in order to check the similarity (in percentage) of 2 given columns.
- The percentage of similarity is computed by dividing the number of similar samples to total samples. `check_similarity` return a number illustrating the similarity of 2 columns.


````python
def check_similarity(data, col_name_1, col_name_2):
    # YOUR CODE HERE
    total_samples = len(data[col_name_1])
    similar_samples = 0

    for i in range(total_samples):
        if data[col_name_1][i] == data[col_name_2][i]:
            similar_samples += 1
    
    return similar_samples * 1.0 / total_samples
    #raise NotImplementedError()
    
````

````python
# TEST
assert round(check_similarity(data_2018, 'Accident Fields_Reference Number', 'Vehicle Fields_Reference Number'),5) == 1.
assert round(check_similarity(data_2018, 'Accident Fields_Reference Number', 'Casualty Fields_Reference Number'),5) == 1.
assert round(check_similarity(data_2018, 'Number of Vehicles', 'Vehicle Number'), 5) == 0.65213
assert round(check_similarity(data_2019, 'Number of Vehicles', 'Vehicle Number'), 5) == 0.67541
````

- In reallife project, you have to convert all columns into their correct datatype before doing anything else. But in this assignment, you are just required to merge `Time (24hr)` and `Accident Date` into column `Full time` with the correct datatype (datetime). After merging, delete `Time (24hr)` and `Accident Date`.
    - The format of `Time (24hr)`: `hhmm`
    - The format of `Accident Date`: `dd/mm/yyyy`

````python
# YOUR CODE HERE
# raise NotImplementedError()

def convertToFullTime(date, time):
    full_time = date + time
    full_time = datetime.datetime.strptime(full_time, "%d/%m/%Y%H%M")
    return full_time

data_18_19['Full time'] = []

for i in range(len(data_18_19['Accident Date'])):
    date = data_18_19['Accident Date'][i]
    time = data_18_19['Time (24hr)'][i]
    data_18_19['Full time'].append(convertToFullTime(date, time))

del data_18_19['Accident Date'], data_18_19['Time (24hr)']
````

### 4.2. Question #2

YOUR ANSWER HERE

- Question: How the accidents are distributed in UK?
- Meaning: Warn people to be careful when participating in traffic in some areas
- Answer:
    - Firstly, I will convert the `Grid Ref` into `Latitude` and `Longitude` using the code from [stackoverflow](https://stackoverflow.com/questions/62019398/fast-conversion-of-easting-and-northing-to-latitude-and-longitude-for-a-large-da)
    - Then, I will use `folium` to visualize the result

````python
# convert easting and northing to latitude and longitude
# source: https://stackoverflow.com/questions/62019398/fast-conversion-of-easting-and-northing-to-latitude-and-longitude-for-a-large-da

import pyproj as proj 
from pyproj import Proj, transform


# convert esting and northing to int
data_18_19['Grid Ref: Easting'] = [int(x) for x in data_18_19['Grid Ref: Easting']]
data_18_19['Grid Ref: Northing'] = [int(x) for x in data_18_19['Grid Ref: Northing']]
 
# convert to latitude and longitude
v84 = Proj(proj="latlong",towgs84="0,0,0",ellps="WGS84")
v36 = Proj(proj="latlong", k=0.9996012717, ellps="airy",
        towgs84="446.448,-125.157,542.060,0.1502,0.2470,0.8421,-20.4894")
vgrid = Proj(init="world:bng")

vlon36, vlat36 = vgrid(data_18_19['Grid Ref: Easting'], data_18_19['Grid Ref: Northing'], inverse=True)
converted = transform(v36, v84, vlon36, vlat36)
latitude = converted[1]
longitude = converted[0]
print(latitude[0], longitude[0])
````

````python
# YOUR CODE HERE
#raise NotImplementedError()

# Plotting each Latitude and Longitude as a point on the UK map using folium
import folium

# calculate the center of the map
center_lat = sum(latitude) / len(latitude)
center_lon = sum(longitude) / len(longitude)

# Create map object at first latitude and longitude
map_uk = folium.Map(location=[center_lat, center_lon], zoom_start=11)

# Add all the points to the map as a circle marker
for lat, lon in zip(latitude, longitude):
    folium.CircleMarker(location=[lat, lon], radius=1.3, color='red', fill_color='red').add_to(map_uk)

# Display the map, map can zoom in and out to see the points clearly
map_uk


````
